# Data Analysis and Visualization on mtcars inbuilt dataset in R

- **Dataset Overview**: The `mtcars` dataset is an inbuilt dataset in R that contains information about various car models. It was originally extracted from a 1974 issue of Motor Trend magazine and is commonly used for data analysis and visualization exercises.

- **Number of Observations**: The dataset consists of 32 rows, each representing a different car model.
- **Variables**: There are 11 variables (columns) in the dataset, including:
  - `mpg`: Miles per gallon (fuel efficiency).
  - `cyl`: Number of cylinders in the engine.
  - `disp`: Displacement (engine size) in cubic inches.
  - `hp`: Horsepower (engine power).
  - `drat`: Rear axle ratio.
  - `wt`: Weight of the car in thousands of pounds.
  - `qsec`: Quarter-mile time (time to accelerate from 0 to 60 mph) in seconds.
  - `vs`: Engine type (0 for V-shaped, 1 for straight).
  - `am`: Transmission type (0 for automatic, 1 for manual).
  - `gear`: Number of forward gears.
  - `carb`: Number of carburetors.

- **Use Cases**: The `mtcars` dataset is often used for various data analysis and visualization tasks, including regression analysis to predict fuel efficiency (`mpg`) based on other vehicle attributes, clustering to group similar car models, and classification to predict transmission type (`am`) based on engine and performance characteristics.

- **Data Source**: This dataset is readily available in R and can be loaded using the `data(mtcars)` command. It serves as a useful example dataset for learning and practicing data analysis techniques in R.

```r
# Loading datset
# This will load the dataset into mtcars variable
data(mtcars)
# Printing dataset
print("Dataset is:")
print(mtcars)
```

```r
# Seeing summary of dataset
print("Summary of Dataset is:")
summary(mtcars)
```

```r
# Seeing first 6 rows of dataset
print("First 6 rows of dataset is:")
head(mtcars)
```

# Data Visualization

```r
# Correlation matrix
# This shows the correlation coefficients between each pair of variables in the dataset, indicating how strongly each variable is related to the others.
cor(mtcars)
```

```r
# Installing and loading the 'corrplot' package
install.packages("corrplot")
library(corrplot)

# Calculate correlation matrix
cor_matrix <- cor(mtcars[, c("mpg", "disp", "hp", "wt", "qsec")])

# Create a correlation heatmap
corrplot(cor_matrix, method = "color", type = "upper", tl.cex = 0.8, title = "Correlation Heatmap")
```

```r
# Scatterplot matrix
# This provides a visual representation of the relationships between each pair of variables in the dataset.
pairs(mtcars, main = "Scatterplot Matrix of mtcars Dataset")
```

```r
# Boxplots by number of cylinders
# This shows the distribution of the miles per gallon variable (mpg) for different numbers of cylinders in the cars.
boxplot(mpg ~ cyl, data = mtcars, main = "Miles per Gallon by Number of Cylinders",
        xlab = "Number of Cylinders", ylab = "Miles per Gallon",col = 'blue')
```

```r
# Histogram of miles per gallon
# This provides a visual representation of the distribution of the miles per gallon variable.
hist(mtcars$mpg, main = "Histogram of Miles per Gallon", xlab = "Miles per Gallon",col = 'darkgreen')
```

```r
# Density plot of miles per gallon
# This provides a smoothed version of the histogram to better visualize the distribution of the miles per gallon variable.
plot(density(mtcars$mpg), main = "Density Plot of Miles per Gallon", xlab = "Miles per Gallon", ylab = "Density",col = "black")
```

```r
# Linear regression of miles per gallon by weight
# This shows the relationship between miles per gallon and weight, and indicates that there is a negative correlation between the two variables.
plot(mtcars$wt, mtcars$mpg, main = "Miles per Gallon by Weight", xlab = "Weight", ylab = "Miles per Gallon", col= "blue", pch = 16, cex = 2)
abline(lm(mpg ~ wt, data = mtcars), col = "red")
```

```r
# ANOVA of miles per gallon by number of cylinders
# This tests whether there is a significant difference in the mean miles per gallon between cars with different numbers of cylinders.
# The analysis shows that the difference is statistically significant, indicating that the number of cylinders is a significant factor in determining the miles per gallon of a car.
fit <- lm(mpg ~ cyl, data = mtcars)
print("Result of Anova test is:")
anova(fit)
```

```r
# We can also use this for classification by creating a target variable

# Create a binary classification target variable
mtcars$mpg_class <- ifelse(mtcars$mpg > median(mtcars$mpg), "High MPG", "Low MPG")

# Load necessary libraries for plotting
library(ggplot2)

# Create a classification plot
ggplot(mtcars, aes(x = wt, y = hp, color = mpg_class)) +
  geom_point(cex = 3) +
  labs(
    title = "Classification Plot of mtcars Dataset for MPG class",
    x = "Weight",
    y = "Horsepower",
    color = "MPG Class"
  )
```

```r
# To predict the transmission type (am) based on engine and performance characteristics in the mtcars dataset

# We need to install this packages before using them
# install.packages('caret')
# install.packages('dplyr')

# Load necessary libraries
library(caret)
library(dplyr)

# Create a binary classification target variable (1 for manual, 0 for automatic)
# Create a new variable 'am' in the 'mtcars' dataset.
# Convert the existing 'am' values (0 and 1) into categorical labels 'Automatic' and 'Manual'.
mtcars <- mtcars %>%
  mutate(am = factor(am, levels = c(0, 1), labels = c("Automatic", "Manual")))


# Split the dataset into training and testing sets
set.seed(123)  # Set a random seed for reproducibility
train_indices <- createDataPartition(mtcars$am, p = 0.7, list = FALSE)
train_data <- mtcars[train_indices, ]
test_data <- mtcars[-train_indices, ]

# Train a logistic regression model
model <- glm(am ~ mpg + disp + hp + drat + wt + qsec + vs + gear + carb, data = train_data, family = binomial)

# Make predictions on the test set
predictions <- predict(model, test_data, type = "response")
predicted_classes <- ifelse(predictions > 0.5, "Manual", "Automatic")

# Evaluate the model
confusion_matrix <- table(predicted = predicted_classes, actual = test_data$am)
accuracy <- sum(diag(confusion_matrix)) / sum(confusion_matrix)
cat("\nAccuracy of model is:", accuracy, "\n")
cat("\nConfusion Matrix of model is:", accuracy, "\n")
print(confusion_matrix)
```

```r
# Visualizing

# Load necessary libraries for plotting
library(ggplot2)

# Create a classification plot
ggplot(mtcars, aes(x = wt, y = hp, color = am)) +
  geom_point(cex = 3) +
  labs(
    title = "Classification Plot of mtcars Dataset for am",
    x = "weight",
    y = "Horsepower",
    color = "am"
  )
```
